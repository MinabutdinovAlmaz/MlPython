```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Сопоставление ключевых точек между двумя изображениями с помощью ORB и Brute-Force Matcher.
# Загрузка двух изображений
image1 = cv2.imread('pic1.jpg', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('pic2.jpg', cv2.IMREAD_GRAYSCALE)

#Приведение размеров изображений к 800 на 600
scaled_image1 = cv2.resize(image1, (800, 600))
scaled_image2 = cv2.resize(image2, (800, 600))
```

```python
# Инициализация ORB
orb = cv2.ORB_create()

# Поиск ключевых точек и дескрипторов
keypoints1, descriptors1 = orb.detectAndCompute(scaled_image1, None)
keypoints2, descriptors2 = orb.detectAndCompute(scaled_image2, None)

# Сопоставление дескрипторов с помощью Brute-Force Matcher
# Для ORB используется норма HAMMING, так как дескрипторы бинарные
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(descriptors1, descriptors2)

# Сортировка сопоставлений по расстоянию (чем меньше расстояние, тем лучше совпадение)
matches = sorted(matches, key=lambda x: x.distance)

# Рисование первых 50 сопоставлений
output_image = cv2.drawMatches(scaled_image1, keypoints1, scaled_image2, keypoints2, matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
# Отображение результата
plt.figure(figsize=(15, 10))
plt.imshow(output_image, cmap='gray')
plt.title('ORB Пересечение Баумана и Чернышевского')
plt.axis('off')
plt.show()

# Вывод информации о сопоставлениях
print(f"Количество сопоставлений: {len(matches)}")
print(f"Лучшее расстояние: {matches[0].distance}")
```

```python
#Сопоставление ключевых точек между двумя изображениями с помощью SIFT

# Инициализация SIFT
sift = cv2.SIFT_create()

# Поиск ключевых точек и дескрипторов
keypoints1, descriptors1 = sift.detectAndCompute(scaled_image1, None)
keypoints2, descriptors2 = sift.detectAndCompute(scaled_image2, None)

# Сопоставление дескрипторов с помощью Brute-Force Matcher
# Для SIFT используется норма L2, так как дескрипторы не бинарные
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = bf.match(descriptors1, descriptors2)

# Сортировка сопоставлений по расстоянию (чем меньше расстояние, тем лучше совпадение)
matches = sorted(matches, key=lambda x: x.distance)

# Рисование первых 50 сопоставлений
output_image = cv2.drawMatches(scaled_image1, keypoints1, scaled_image2, keypoints2, matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
# Отображение результата
plt.figure(figsize=(15, 10))
plt.imshow(output_image, cmap='gray')
plt.title('SIFT Пересечение Баумана и Чернышевского')
plt.axis('off')
plt.show()

# Вывод информации о сопоставлениях
print(f"Количество сопоставлений: {len(matches)}")
print(f"Лучшее расстояние: {matches[0].distance}")
```
