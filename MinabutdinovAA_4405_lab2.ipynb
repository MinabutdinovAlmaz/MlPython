```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Загрузка данных CIFAR-10
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
batch_size = 64 #Размер мини-выборки
```

```python
#Генератор Аугментации
datagen = ImageDataGenerator(
    rescale=1./255,#Нормализация данных
    rotation_range=30, #Поворачиваем изображения при генерации выборки
    width_shift_range=0.2, #Двигаем изображения по ширине при генерации выборки
    height_shift_range=0.2, #Двигаем изображения по высоте при генерации выборки
    brightness_range=[0.5, 1.5], #изменение яркости
    zoom_range=0.2, #Зумируем изображения при генерации выборки
    horizontal_flip=True, #Включаем отзеркаливание изображений
    fill_mode='nearest' #Заполнение пикселей вне границ ввода
)
```

```python
random_number = random.randint(0, 50000)
plt.axis('off')
plt.title("Исходное изображение")
plt.imshow(x_train[random_number]) #Вывод исходного изображения
#Аугментация
x = tf.keras.utils.img_to_array(x_train[random_number])
x = x.reshape((1,) + x.shape)
plt.figure(figsize=(15, 3))
i=0
for batch in datagen.flow(x, batch_size=1):
    imgplot = plt.subplot(1, 5, i + 1)  # 1 строка, 5 колонок
    plt.imshow(tf.keras.utils.img_to_array(batch[0]))
    plt.axis('off')
    i += 1
    if i >= 5:  # Лимитируем количество отображаемых изображений до 5
        break
plt.suptitle("Аугментированные изображения")
plt.show()
```
